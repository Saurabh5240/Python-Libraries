#Seaborn
- Data visualization library

```python
#importing libraries
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Note - seaborn has some bulit in datasets for practice

```python
#total bill vs tip dataset
tips = sns.load_dataset('tips')
tips.head()
```

```python
#setting a theme
sns.set_theme()
```

```python
#visualize the data
sns.relplot(data = tips, x = 'total_bill',y='tip',col='time',hue = 'smoker', style= 'smoker',size = 'size')
```

```python
#loading iris dataset
iris = sns.load_dataset("iris")
iris.head()
```

```python
#Using the scatter plot
sns.scatterplot(data=iris,x= 'sepal_length',y='petal_length',hue='species')
```

```python
#importing titanic dataset
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
#Count plot
sns.countplot(data= titanic,x='class',hue= 'sex')
```

```python
#Bar chart
sns.barplot(data=titanic,x='sex',y='survived',hue='class')
```

```python
#importing housing dataset
housing = pd.read_csv("HousingData.csv")
housing.head()
```

```python
#Distribution plot
sns.displot(data=housing,x='PRICE') #or use sns.distplot(housing['PRICE]) or use sns.histplot
```

```python
sns.distplot(housing['PRICE'])
```

```python
#Correlation and HeatMap
correlation = housing.corr()
plt.figure(figsize=(10,10))
sns.heatmap(correlation,cbar=True,square=True,fmt='.1f',annot=True,annot_kws={'size':8},cmap='Greens')
```
