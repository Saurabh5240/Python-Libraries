# MatPlotLib
- used to ctreate plots

```python
#importing matplotlib
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
```

```python
print(x,"\n",y,"\n",z)
plt.plot(z,y)       #something new sin and cos we get a circle on a axes
plt.show()
```

```python
#To plot the curve
plt.plot(x,y)
plt.plot(x,z)
plt.show()
```

```python
#To add labels to the graph
plt.plot(x,z)     
plt.xlabel("Angle")
 
plt.ylabel("Cosine value")   #since no zalbel is present only x and y label is used
plt.title("Cosine Wave")
plt.show()
```

```python
#plot a parabola

x = np.linspace(-10,10,20)
y = x**2
plt.plot(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Parabola")
plt.show()
```

```python
#plot the graphs with dots of desired symbols and colours
plt.plot(x,y,"g.")   #here r denotes red and * is symbol used eg - g for green 
plt.show()
```

```python
#An example
x = np.linspace(0,10,100)
plt.plot(x, np.sin(x), "g-")
plt.plot(x,np.cos(x),"r.")
plt.show()
```

BAR PLOT

```python
fig = plt.figure()                                   # to create a empty plot
ax = fig.add_axes([0,0,1,1])                         #a rectangle of 4 lenth sequence [left,bottom,width,height]
language = ["Hindi","English","Spanish","German","Urdu"]
people  = [100,200,34,66,98]
ax.bar(language,people)
plt.xlabel("Languages")
plt.ylabel("No of People")
plt.show()
```

Pie chart

```python
fig1 = plt.figure()
ax = fig1.add_axes([0,0,1,1])
language = ["Hindi","English","Spanish","German","Urdu"]
people  = [100,200,34,66,98]
ax.pie(people, labels = language, autopct = '%1.1f%%')
plt.show()
```

Scatter Plot

```python
x = np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
fig2 = plt.figure()
ax = fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color = 'r')
ax.scatter(x,z,color= 'g')
plt.show()
```

```python
#To create a 3d scatter plot

fig3 = plt.figure()
ax = plt.axes(projection ='3d')
x = 10* np.linspace(0,10,100)
y = np.sin(x)
z = np.cos(x)
ax.scatter(x,y,x,c=z,cmap = 'Blues')
plt.show()
```
