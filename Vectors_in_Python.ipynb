Dot Product

```python
a = np.asarray([2,3,4])
b = np.asarray([4,4,4])

a_dot_b = np.dot(a,b)
print(a_dot_b)
```

Cross product

```python
a = np.asarray([2,3])
b = np.asarray([4,4])
a_cross_b = np.cross(a,b)
print(a_cross_b)   #since other values are 0 so it shoes -4 only , the answer is [0 0 -4]
```

```python
a = np.asarray([2,3,2])
b = np.asarray([4,4,3])
a_cross_b = np.cross(a,b)
print(a_cross_b)
```

Projection of l on m vector--*italicized text*

```python
a = np.array([2, 5])

v = np.array([8, -6])

# magnitude of "v" vector
magnitude_of_v = np.sqrt(np.sum(v**2))

proj_of_a_on_v = (np.dot(a,v)/magnitude_of_v**2)*v

print('Projection of a vector on v vector = ', proj_of_a_on_v)
```

```python
a = np.array([23, 45, 62])

v = np.array([45, 82, 67])

# magnitude of "v" vector
magnitude_of_v = np.sqrt(np.sum(v**2))

proj_of_a_on_v = (np.dot(a,v)/magnitude_of_v**2)*v

print('Projection of a vector on v vector = ', proj_of_a_on_v)
```
